```python
loader = TextLoader("/content/merged_resume.txt")

documents = loader.load()
text_splitter = CharacterTextSplitter(separator=" ", chunk_size=2000, chunk_overlap=0)
```

```python
docs = text_splitter.split_documents(documents)
```

```python
print(f'Всего фрагментов: {len(docs)}')
```

```python
# первый чанк
# print(docs[0])
page_content = docs[0].page_content
print(page_content, '\n')

# длина первого чанка
print('Длина чанка', len(page_content))
```

**Кодирование (Эмбеддинг)**

```python
# Инициализирум модель эмбеддингов
embeddings = OpenAIEmbeddings()

# Создадим индексную базу из разделенных фрагментов текста
db = FAISS.from_documents(docs, embeddings)
```

```python
query = "Опыт работы для специалиста владеющего Python, C++, TensorFlow, Keras, Компьютерным зрением."
results = db.similarity_search(query, k=4)

print("**отрезок №1**\n", results[0].page_content)
print("\n**отрезок №2**\n", results[1].page_content)
print("\n**отрезок №3**\n", results[2].page_content)
print("\n**отрезок №4**\n", results[3].page_content)
```

Роль chatGPT

```python
system = """Ты помощник в составлении резюме для кандидатов на позицию специалиста или дата-сайентиста (data scientist) в области искусственного интеллекта.
Ты профессионал в своей области. Тебе предоставят инструкцию по составлению резюме и попросят помочь кандидату составить резюме, опираясь на эту инструкцию.

Твоя цель: Опираясь на предоставленную инструкцию, составить качественное и грамотное резюме или его часть на русском языке, не превышая 5000 слов.
Отвечай строго по запросу, если тебя просят составить только часть резюме, то только эту часть и формируй без лишней информации.
Если в инструкции отсутствуют необходимые данные, ты можешь добавить их самостоятельно, с учетом контекста.
Не упоминай предоставленную тебе инструкцию в ответе, а используй ее только для составления резюме.
Отвечай только на запросы, связанные с составлением резюме, и не включай лишнюю информацию, такую как примечания, заключения или выводы."""
```

Функция для работы с chatGPT

```python
# функция для работы с chatGPT
def generate_answer(question):
    # Выполните поиск документов с использованием db.similarity_search
    docs = db.similarity_search(question, k=6)

    # Сформируйте сообщение, которое будет передано в OpenAI
    message_content = re.sub(r'\n{2}', ' ', '\n '.join([f'\nОтрывок документа №{i+1}\n=====================' + doc.page_content + '\n' for i, doc in enumerate(docs)]))
    # print(message_content)
    messages = [
        {"role": "system", "content": system},
        {"role": "user", "content": f"Документ с информацией для ответа клиенту: {message_content}\n\nВопрос клиента: \n{question}"}
    ]

    # Запросите ответ у модели GPT-3
    completion = openai.ChatCompletion.create(
        model="gpt-3.5-turbo-16k",
        messages=messages,
        temperature=0
    )

    # Верните ответ
    return completion.choices[0].message.content
```

Взаимодействие с chatGPT

```python
question = """
Составь полное резюме для молодого специалиста без опыта работы в области искусственного интеллекта, но владеющего Python, TensorFlow, Keras, PyTorch, scikit-learn, pandas, NumPy, Matplotlib.
"""
mess = generate_answer(question)
print(mess.replace("\\n", "\n"))
```
