```python
# А это для модификации многих строк
# Функция для внесения ошибок в текст
def introduce_typo(word):
    if len(word) <= 1:
        return word
    else:
        index = random.randint(0, len(word) - 1)
        new_char = chr(random.randint(ord('а'), ord('я')))
        return word[:index] + new_char + word[index+1:]

# Функция для внесения ошибок и перестановки слов в тексте
def introduce_errors(text, error_rate=0.4, shuffle_rate=0.4):  # error_rate  - коэфф. вноса ошибок, shuffle_rate - коэфф. перемешивания слов
    words = text.split()

    for i in range(len(words)):
        # Внесение ошибок
        if random.random() < error_rate:
            words[i] = introduce_typo(words[i])

        # Перестановка слов
        if random.random() < shuffle_rate and i < len(words) - 1:
            j = i + 1
            words[i], words[j] = words[j], words[i]

    return ' '.join(words)

# Чтение данных из .csv файла
df = pd.read_csv('/content/800_fake_rez_zvf.csv', delimiter=';', encoding='Windows-1251')

# Применение функции introduce_errors к столбцу 'Обо мне'
df['Обо_мне_err'] = df['Обо мне'].apply(introduce_errors)

# Сохранение измененных данных в новый .csv файл
# df.to_csv('новый_файл.csv', index=False, sep=';')
# Сохранение измененных данных в новый .csv файл с кодировкой 'Windows-1251'
df.to_csv('новый_файл.csv', index=False, sep=';', encoding='Windows-1251')
df
```
